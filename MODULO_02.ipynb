## Filtros (seleção booleana)

Explicar e exemplificar como filtrar linhas com base em condições.


```python
%%markdown
## Filtros (Seleção Booleana)
```

```python
%%markdown
A filtragem booleana no pandas é uma técnica poderosa para selecionar subconjuntos de linhas em um DataFrame com base em condições específicas. Ela envolve a criação de uma **série booleana** (uma série de valores `True` e `False`) onde cada valor corresponde a uma linha no DataFrame original. `True` indica que a linha satisfaz a condição, e `False` indica que não. Ao passar essa série booleana para os colchetes do DataFrame, apenas as linhas onde a série booleana é `True` são selecionadas.

Vamos ver um exemplo filtrando as linhas onde a idade é maior que 16:
```

**Reasoning**:
Demonstrate filtering rows where the 'idade' column is greater than 16 and display the head of the resulting DataFrame.



```python
filtro_idade = df['idade'] > 16
df_idade_maior_que_16 = df[filtro_idade]
display(df_idade_maior_que_16.head())
```

```python
%%markdown
Agora, vamos filtrar o DataFrame para selecionar apenas as linhas onde a coluna 'descricao' é igual a 'Cancelado'.
```

```python
filtro_cancelado = df['descricao'] == 'Cancelado'
df_cancelado = df[filtro_cancelado]
display(df_cancelado.head())
```

```python
%%markdown
É possível combinar múltiplas condições de filtragem usando os operadores lógicos do pandas:

*   `&` para a operação lógica E (AND)
*   `|` para a operação lógica OU (OR)
*   `~` para a negação lógica (NOT)

É importante agrupar cada condição entre parênteses para garantir a precedência correta dos operadores.

Vamos filtrar as linhas onde a idade é maior que 16 **E** a descrição é igual a 'Cancelado'.
```

Além da combinação com o operador AND (`&`), podemos usar os operadores lógicos OR (`|`) para selecionar linhas que satisfaçam **qualquer uma** das condições e NOT (`~`) para selecionar linhas que **não** satisfaçam uma condição.

Vamos filtrar as linhas onde a descrição é 'Cancelado' **OU** a idade é maior que 20.

```python
filtro_or = (df['descricao'] == 'Cancelado') | (df['idade'] > 20)
df_filtrado_or = df[filtro_or]
display(df_filtrado_or.head())
```

Agora, vamos filtrar as linhas onde a descrição **NÃO** é 'Matriculado'.

```python
filtro_not = ~(df['descricao'] == 'Matriculado')
df_filtrado_not = df[filtro_not]
display(df_filtrado_not.head())
```

## Slices (seleção por índice)

Mostrar como selecionar linhas e colunas usando slices baseados em índices.


```python
%%markdown
## Slices (Seleção por Índice)
```

```python
%%markdown
Slicing em pandas permite selecionar um intervalo contínuo de linhas ou colunas de um DataFrame usando a notação de colchetes `[]` com índices inteiros. É importante notar que, ao contrário da seleção por rótulo (`.loc`), o slicing com colchetes no nível superior do DataFrame (`df[]`) opera **exclusivamente** nos rótulos das linhas (índices). No entanto, quando usado em conjunto com `.iloc`, o slicing opera em índices inteiros e é **exclusivo** no limite superior, tanto para linhas quanto para colunas.
```

```python
# Selecionando as primeiras 5 linhas usando slicing
# Ao usar slicing diretamente no DataFrame (df[]), o limite superior é inclusivo para rótulos.
# No entanto, com iloc (que usa indices inteiros), o limite superior é exclusivo.
# Como o índice padrão é numérico e sequencial, fatiar [0:5] selecionará as linhas com índice 0, 1, 2, 3, 4.
primeiras_5_linhas = df[0:5]
display(primeiras_5_linhas)
```

```python
# Selecionando linhas e colunas usando slicing combinado com .iloc
# .iloc usa índices inteiros e o limite superior é exclusivo tanto para linhas quanto para colunas.
# Seleciona linhas com índices 5, 6, 7, 8, 9.
# Seleciona colunas com índices 2, 3, 4.
selecao_combinada_iloc = df.iloc[5:10, 2:5]
display(selecao_combinada_iloc)
```

```python
%%markdown
É crucial entender a diferença no comportamento do slicing entre usar os colchetes `[]` diretamente no DataFrame e usar `.iloc[]`:

*   **Slicing direto com `[]` (para seleção de linhas):** Quando você usa `df[start:stop]`, o `start` é inclusivo e o `stop` é **inclusivo** se o índice for baseado em rótulos (como strings ou datas) ou **exclusivo** se o índice for baseado em inteiros padrão (0, 1, 2...). No nosso caso, com o índice padrão, `df[0:5]` incluiu as linhas de índice 0 a 4.
*   **Slicing com `.iloc[]` (para seleção por posição):** Quando você usa `df.iloc[start:stop, start_col:stop_col]`, tanto o `start` quanto o `start_col` são inclusivos, mas o `stop` e o `stop_col` são **exclusivos**. Ou seja, `df.iloc[5:10, 2:5]` selecionou as linhas com posições de 5 a 9 e as colunas com posições de 2 a 4.

Essa diferença pode ser uma fonte comum de erros, então sempre verifique se você está usando `.loc` (seleção por rótulo, limite superior inclusivo) ou `.iloc` (seleção por posição, limite superior exclusivo) para garantir o comportamento esperado.
```

## Iloc

Explicar e demonstrar o uso de `iloc` para selecionar dados com base em posições inteiras.


```python
%%markdown
## iloc (Seleção por Posição Inteira)
```

```python
%%markdown
O `iloc` é o método principal no pandas para seleção de dados baseada **exclusivamente** em posições inteiras das linhas e colunas. As posições são baseadas em zero, semelhante a listas ou arrays NumPy.

Você pode usar `iloc` para selecionar:
- Uma única linha por sua posição inteira.
- Múltiplas linhas por uma lista ou slice de posições inteiras.
- Uma única coluna por sua posição inteira.
- Múltiplas colunas por uma lista ou slice de posições inteiras.
- Combinações de linhas e colunas usando posições inteiras.

A sintaxe geral é `df.iloc[linhas, colunas]`, onde `linhas` e `colunas` podem ser um único inteiro, uma lista de inteiros ou um slice de inteiros. É importante lembrar que, com `iloc`, o limite superior dos slices é **exclusivo**.
```

```python
# Demonstrando a seleção de uma única linha usando iloc (a primeira linha, posição 0)
primeira_linha_iloc = df.iloc[0]
display(primeira_linha_iloc)
```

```python
# Demonstrando a seleção de múltiplas linhas usando uma lista de posições com iloc
# Selecionando as linhas nas posições 0, 2 e 4
multiplas_linhas_iloc = df.iloc[[0, 2, 4]]
display(multiplas_linhas_iloc)
```

```python
# Demonstrando a seleção de linhas e colunas específicas usando iloc com slices
# Selecionando as primeiras 10 linhas (posições 0 a 9) e as colunas da posição 5 à 10 (posições 5 a 9)
selecao_slices_iloc = df.iloc[0:10, 5:10]
display(selecao_slices_iloc)
```

```python
# Demonstrando a seleção de linhas específicas e todas as colunas usando iloc
# Selecionando as linhas nas posições 0, 2 e 4 e todas as colunas
selecao_linhas_especificas_iloc = df.iloc[[0, 2, 4], :]
display(selecao_linhas_especificas_iloc)
```

```python
# Demonstrando a seleção de todas as linhas e colunas específicas usando iloc
# Selecionando todas as linhas e as colunas nas posições 0, 5 e 10
selecao_colunas_especificas_iloc = df.iloc[:, [0, 5, 10]]
display(selecao_colunas_especificas_iloc)
```

## Comparando `iloc`, `loc` e slices

Comparar e contrastar os métodos `iloc`, `loc` e o slicing direto em pandas, destacando suas diferenças e quando usar cada um.


```python
%%markdown
## Comparando `iloc`, `loc` e Slices
```

```python
%%markdown
Compreender as diferenças entre `iloc`, `loc` e o slicing direto com `[]` é fundamental para manipular dados de forma eficaz no pandas. Embora todos sirvam para selecionar subconjuntos de dados, eles o fazem de maneiras distintas:

*   **`iloc`**: Usado para seleção baseada **exclusivamente em posições inteiras**. Pense nele como acessar elementos em uma lista ou array NumPy. O índice de linha e coluna começa em 0. Ao usar slices com `iloc` (por exemplo, `df.iloc[0:5, 1:3]`), o limite superior do slice é **exclusivo**. É ideal para automação onde você sabe a posição dos dados que precisa.

*   **`loc`**: Usado para seleção baseada **exclusivamente em rótulos** (nomes de índice de linha e nomes de coluna). Você precisa saber os nomes exatos das linhas e colunas que deseja acessar. Ao usar slices com `loc` (por exemplo, `df.loc[0:5, 'idade':'descricao']`), o limite superior do slice é **inclusivo**. É mais legível e menos propenso a erros quando seus índices de linha ou nomes de coluna têm significado.

*   **Slicing direto com `[]` no DataFrame**: Quando usado para selecionar linhas (por exemplo, `df[0:5]`), ele opera nos **rótulos do índice de linha**. Se o índice for o padrão baseado em inteiros (0, 1, 2...), o comportamento do slice (`[start:stop]`) será semelhante ao slicing de lista/NumPy, com o limite superior **exclusivo**. Se o índice for baseado em rótulos não inteiros (como strings ou datas), o limite superior do slice será **inclusivo**. Para seleção de colunas, `df['coluna']` ou `df[['col1', 'col2']]` usam rótulos. A sintaxe `df[condição booleana]` é usada para filtragem.

A principal distinção reside em **como** você especifica os dados que deseja: por sua posição (com `iloc`) ou por seu nome/rótulo (com `loc` e, em certos casos, com slicing direto). O comportamento do limite superior nos slices também difere entre `iloc` (exclusivo) e `loc` (inclusivo).
```

```python
# Exemplo: Selecionando as primeiras 3 linhas e as primeiras 3 colunas

print("Usando iloc (seleção por posição, limite superior exclusivo):")
# Seleciona linhas nas posições 0, 1, 2 e colunas nas posições 0, 1, 2
display(df.iloc[0:3, 0:3])

print("\nUsando loc (seleção por rótulo, limite superior inclusivo):")
# Seleciona linhas com rótulos 0, 1, 2 e colunas com rótulos 'LnguaPortuguesaeLiteraturaI90H' a 'LnguaPortuguesaeLiteraturaI90H_freq' (inclusive)
# Como o índice é padrão (0, 1, 2...), o slice de linhas se comporta como iloc para este caso.
display(df.loc[0:2, 'LnguaPortuguesaeLiteraturaI90H':'LnguaPortuguesaeLiteraturaI90H_freq'])

print("\nUsando slicing direto no DataFrame (seleção de linhas por rótulo/índice, colunas por rótulo):")
# Seleciona linhas com rótulos 0, 1, 2 (com índice padrão, comporta-se como iloc para linhas)
# Seleciona colunas 'LnguaPortuguesaeLiteraturaI90H', 'LnguaPortuguesaeLiteraturaI90H_dependencia', 'LnguaPortuguesaeLiteraturaI90H_freq' por lista de rótulos
display(df[0:3][['LnguaPortuguesaeLiteraturaI90H', 'LnguaPortuguesaeLiteraturaI90H_dependencia', 'LnguaPortuguesaeLiteraturaI90H_freq']])
```

```python
%%markdown
### Quando usar cada método:

*   **Use `iloc`**: Quando você precisa selecionar dados com base em sua posição numérica exata. Isso é comum em loops, automação, ou quando você não tem controle sobre os rótulos do índice (por exemplo, após redefinir o índice). Lembre-se do limite superior **exclusivo** em slices.
*   **Use `loc`**: Quando você precisa selecionar dados usando seus rótulos de linha e/ou nomes de coluna. É a abordagem mais legível quando os rótulos têm significado e é menos frágil a mudanças na ordem das linhas ou colunas. Lembre-se do limite superior **inclusivo** em slices.
*   **Use Slicing direto com `[]`**: Para selecionar linhas por posição/rótulo (com o comportamento de limite superior dependente do tipo de índice) ou para selecionar colunas por rótulo (`df[['col1', 'col2']]`). Também é usado para filtragem booleana (`df[condição]`).

Em resumo, `iloc` para posições, `loc` para rótulos e slicing direto para conveniência na seleção de linhas/colunas inteiras ou filtragem. Escolher o método correto torna seu código mais claro, eficiente e menos propenso a erros.
```

Apresentar brevemente outros métodos comuns como `dropna()`, `fillna()`, `groupby()`, `sort_values()`, etc.

```python
%%markdown
O pandas oferece uma vasta gama de métodos para manipular e transformar dados. Além da seleção e filtragem, alguns dos métodos mais frequentemente utilizados incluem:

*   **`dropna()`**: Este método é usado para **remover linhas ou colunas que contêm valores ausentes** (NaN). É uma etapa comum na limpeza de dados para lidar com dados incompletos.
*   **`fillna()`**: Permite **preencher valores ausentes** com um valor específico, um método estatístico (como a média ou mediana), ou usando métodos de preenchimento como forward fill (`ffill`) ou backward fill (`bfill`).
*   **`groupby()`**: Um método poderoso para **agrupar linhas com base em um ou mais valores de coluna**. Frequentemente usado em conjunto com funções de agregação (como `sum()`, `mean()`, `count()`) para analisar dados por categoria.
*   **`sort_values()`**: Usado para **ordenar um DataFrame pelas colunas especificadas**. Permite ordenar em ordem crescente ou decrescente.
*   **`merge()`**: Utilizado para **combinar dois DataFrames com base em colunas chave comuns**, similar a operações JOIN em bancos de dados.
*   **`apply()`**: Um método flexível para **aplicar uma função** ao longo de um eixo de um DataFrame (linhas ou colunas). É útil para realizar operações personalizadas que não são diretamente suportadas por métodos pandas embutidos.
```

## Demonstração de `groupby()`

Demonstração de `groupby()`


```python
%%markdown
## Agrupando Dados com `groupby()`

O método `groupby()` no pandas é uma ferramenta essencial para **dividir um DataFrame em grupos** com base nos valores de uma ou mais colunas. Uma vez que os dados são agrupados, você pode aplicar funções de **agregação** a cada grupo separadamente. Isso permite realizar análises sumárias por categoria, como calcular a média, a soma, a contagem, o valor mínimo ou máximo para cada grupo.

A ideia principal por trás de `groupby()` segue um processo de três etapas:

1.  **Splitting (Divisão)**: O DataFrame é dividido em grupos menores com base na(s) coluna(s) especificada(s).
2.  **Applying (Aplicação)**: Uma função é aplicada a cada grupo. Esta pode ser uma função de agregação, uma função de transformação, ou uma função de filtragem.
3.  **Combining (Combinação)**: Os resultados da aplicação da função em cada grupo são combinados em uma estrutura de dados resultante, geralmente um Series ou DataFrame.

Os parâmetros mais comuns de `groupby()` incluem:

*   O nome da coluna (ou uma lista de nomes de colunas) para agrupar.
*   O `axis` ao longo do qual agrupar (0 para linhas, 1 para colunas, embora agrupar por colunas seja menos comum).

Após agrupar, você seleciona a(s) coluna(s) nas quais deseja aplicar a agregação e chama a função de agregação desejada (por exemplo, `.mean()`, `.sum()`, `.count()`, `.size()`, `.min()`, `.max()`).
```

```python
# Agrupando o DataFrame pela coluna 'descricao' e calculando a média da coluna 'idade'
idade_media_por_descricao = df.groupby('descricao')['idade'].mean()
display(idade_media_por_descricao)
```

Após usar `groupby()` e realizar agregações, o resultado frequentemente tem as colunas de agrupamento como o índice do novo DataFrame ou Series. O método `reset_index()` é útil para **converter o índice de volta em uma coluna regular** no DataFrame, tornando-o mais fácil de trabalhar em algumas situações.

```python
# Agrupando o DataFrame pela coluna 'descricao' e calculando o tamanho do grupo e a média de 'qnt_salarios'
agregacao_multipla = df.groupby('descricao').agg(
    total_alunos=('descricao', 'size'),
    media_salarios=('qnt_salarios', 'mean')
)
display(agregacao_multipla)
```

```python
# Demonstrando o uso de reset_index()
# Transforma o índice (que é a coluna 'descricao' após o groupby) em uma coluna normal.
agregacao_resetada = agregacao_multipla.reset_index()
display(agregacao_resetada)
```

Demonstração de `apply()` com lambda


```python
%%markdown
## Aplicando Funções com `apply()` e Lambda

O método `apply()` no pandas é uma forma flexível de **aplicar uma função ao longo de um eixo** de um DataFrame ou Series. Isso significa que você pode usar `apply()` para operar em cada coluna (axis=0, padrão) ou em cada linha (axis=1).

Uma combinação poderosa é usar `apply()` com **funções lambda**. Funções lambda são pequenas funções anônimas que você pode definir e usar em uma única linha. Elas são ideais para operações simples que não justificam a definição de uma função completa usando `def`.

Ao usar `apply()` com uma função lambda:

*   **`axis=0` (padrão)**: A função lambda será aplicada a cada **coluna** individualmente. A entrada para a função lambda será uma Series representando a coluna.
*   **`axis=1`**: A função lambda será aplicada a cada **linha** individualmente. A entrada para a função lambda será uma Series representando a linha.

O `apply()` é muito útil para criar novas colunas baseadas em lógica complexa que envolve os valores de outras colunas, ou para realizar transformações específicas em subconjuntos de dados.
```

**Reasoning**:
Demonstrate using apply() with a lambda function to create a new column based on a condition applied to another column (e.g., categorizing age).



```python
# Criando uma nova coluna 'faixa_idade' usando apply() com lambda
# A função lambda verifica o valor da coluna 'idade' para cada linha (axis=1 é implícito ao operar em uma coluna)
# e atribui uma categoria de idade.
df['faixa_idade'] = df['idade'].apply(lambda x: 'Jovem' if x < 18 else ('Adulto' if x >= 18 and x < 60 else 'Idoso'))

display(df[['idade', 'faixa_idade']].head())
```

```python
# Demonstrando apply() com lambda em linhas (axis=1)
# Criando uma nova coluna 'media_notas' que calcula a média de duas colunas de notas para cada aluno
# É necessário lidar com valores ausentes (NaN) antes de calcular a média, por exemplo, substituindo-os por 0 para a média.
# Para um cálculo de média que ignore NaNs, podemos usar o método .mean() na Series de cada linha.
df['media_notas'] = df.apply(
    lambda row: (row['LnguaPortuguesaeLiteraturaI90H'] + row['MatemticaI120H']) / 2 if pd.notna(row['LnguaPortuguesaeLiteraturaI90H']) and pd.notna(row['MatemticaI120H']) else None,
    axis=1
)

# Alternativamente, usando .mean() na Series da linha para calcular a média ignorando NaNs
df['media_notas_ignoring_nan'] = df[['LnguaPortuguesaeLiteraturaI90H', 'MatemticaI120H']].apply(lambda x: x.mean(), axis=1)


display(df[['LnguaPortuguesaeLiteraturaI90H', 'MatemticaI120H', 'media_notas', 'media_notas_ignoring_nan']].head())
```

## Demonstração de `merge()`

Explicar e exemplificar como combinar os dois DataFrames criados.


```python
%%markdown
## Combinando DataFrames com `merge()`

O método `merge()` no pandas é usado para **combinar dois DataFrames** com base em uma ou mais colunas chave comuns. Ele funciona de forma muito semelhante às operações de JOIN em bancos de dados relacionais. O `merge()` é essencial quando você precisa integrar informações de diferentes fontes de dados que compartilham um identificador comum.

Os parâmetros principais de `merge()` incluem:

*   **`left`**: O DataFrame à esquerda na operação de merge.
*   **`right`**: O DataFrame à direita na operação de merge.
*   **`on`**: O nome da coluna (ou uma lista de nomes de colunas) para usar como chaves de junção. Se as colunas de junção tiverem nomes diferentes nos DataFrames esquerdo e direito, você pode usar `left_on` e `right_on` separadamente.
*   **`how`**: Especifica o tipo de junção a ser realizada. Os tipos mais comuns são:
    *   **`inner` (padrão)**: Retorna apenas as linhas onde as chaves de junção estão presentes em **ambos** os DataFrames (interseção).
    *   **`outer`**: Retorna todas as linhas de **ambos** os DataFrames, preenchendo com `NaN` onde as chaves não correspondem (união).
    *   **`left`**: Retorna todas as linhas do DataFrame **esquerdo** e as linhas correspondentes do DataFrame direito. Preenche com `NaN` para chaves do lado esquerdo que não têm correspondência no lado direito.
    *   **`right`**: Retorna todas as linhas do DataFrame **direito** e as linhas correspondentes do DataFrame esquerdo. Preenche com `NaN` para chaves do lado direito que não têm correspondência no lado esquerdo.

Vamos demonstrar diferentes tipos de merges usando os DataFrames `df_alunos` e `df_notas` que criamos.
```

**Reasoning**:
Demonstrate an inner merge between df_alunos and df_notas on the 'id_aluno' column and display the head of the resulting DataFrame.



```python
# Demonstrando um merge interno (inner merge)
# Retorna apenas as linhas onde o id_aluno existe tanto em df_alunos quanto em df_notas.
df_merged_inner = pd.merge(left=df_alunos, right=df_notas, on='id_aluno', how='inner')
display(df_merged_inner.head(10))
```

```python
# Demonstrando um merge à esquerda (left merge)
# Retorna todas as linhas de df_alunos e as linhas correspondentes de df_notas.
# Alunos em df_alunos sem correspondência em df_notas terão NaN nas colunas de notas.
df_merged_left = pd.merge(left=df_alunos, right=df_notas, on='id_aluno', how='left')
display(df_merged_left.head(10))
```

```python
# Demonstrando um merge à direita (right merge)
# Retorna todas as linhas de df_notas e as linhas correspondentes de df_alunos.
# Notas em df_notas sem correspondência em df_alunos terão NaN nas colunas de alunos.
df_merged_right = pd.merge(left=df_alunos, right=df_notas, on='id_aluno', how='right')
display(df_merged_right.head(10))
```

```python
# Demonstrando um merge externo (outer merge)
# Retorna todas as linhas de ambos os DataFrames.
# Linhas sem correspondência em um dos DataFrames terão NaN nas colunas do outro DataFrame.
df_merged_outer = pd.merge(left=df_alunos, right=df_notas, on='id_aluno', how='outer')
display(df_merged_outer.head(10))
```

## como importar dados de arquivos no github

```python
import pandas as pd

# URL do arquivo JSON no GitHub
url = "https://raw.githubusercontent.com/tmedeirosb/modelo_ifrn_integrado/master/versao_2/dados/dados_workflow_ivan.json"

# Importar usando pandas
try:
    df_pandas = pd.read_json(url)
    print("DataFrame pandas:")
    display(df_pandas.head())
except Exception as e:
    print(f"Erro ao importar com pandas: {e}")
```

## Como salvar dados no google drive

```python
from google.colab import drive
import os

# Desmontar o Google Drive se j estiver montado
if os.path.exists('/content/drive'):
    try:
        drive.flush_and_unmount()
        print("Google Drive desmontado com sucesso.")
    except Exception as e:
        print(f"Erro ao desmontar o Google Drive: {e}")

# Montar o Google Drive
try:
    drive.mount('/content/drive')
    print("Google Drive montado com sucesso.")
except Exception as e:
    print(f"Erro ao montar o Google Drive: {e}")

# Definir o caminho para salvar o arquivo CSV
# Voc pode ajustar o caminho conforme necessrio.
# Este exemplo salva em um subdiretrio especfico no seu Drive.
output_dir = '/content/drive/MyDrive/EAD/DISCIPLINAS/Analise de dados /COLAB/MODULO_01'
output_path = os.path.join(output_dir, 'dados_workflow_ivan.csv')

# Criar o diretrio de sada se no existir
try:
    os.makedirs(output_dir, exist_ok=True)
    print(f"Diretrio de sada verificado/criado: {output_dir}")
except Exception as e:
    print(f"Erro ao criar o diretrio de sada: {e}")


# Salvar o DataFrame como CSV
try:
    df_pandas.to_csv(output_path, index=False)
    print(f"DataFrame salvo com sucesso em: {output_path}")
except Exception as e:
    print(f"Erro ao salvar o DataFrame como CSV: {e}")
```

## Como acessar dados do google drive

```python
import os

# O caminho para o arquivo CSV salvo
# Certifique-se de que o Google Drive ainda est montado e que o caminho est correto
output_dir = '/content/drive/MyDrive/EAD/DISCIPLINAS/Analise de dados /COLAB/MODULO_01'
output_path = os.path.join(output_dir, 'dados_workflow_ivan.csv')

# Importar o arquivo CSV para um novo DataFrame
try:
    df_from_csv = pd.read_csv(output_path)
    print("Novo DataFrame importado do CSV:")
    display(df_from_csv.head())
except FileNotFoundError:
    print(f"Erro: O arquivo no foi encontrado em {output_path}. Certifique-se de que o Google Drive est montado e o caminho est correto.")
except Exception as e:
    print(f"Erro ao importar o arquivo CSV: {e}")
```

```python
#https://github.com/koaning/drawdata

!pip install -q drawdata

from drawdata import ScatterWidget

widget = ScatterWidget()
widget


```

```python
# Get the drawn data as a list of dictionaries
#widget.data

# Get the drawn data as a dataframe
widget.data_as_pandas
```

```python
from google.colab import files

# Obter os dados do widget como um DataFrame pandas
try:
    df_widget_data = widget.data_as_pandas
    print("DataFrame com dados do widget:")
    display(df_widget_data.head())

    # Salvar o DataFrame como CSV em um arquivo temporrio
    csv_file = 'widget_data.csv'
    df_widget_data.to_csv(csv_file, index=False)

    # Fornecer um link para download do arquivo
    files.download(csv_file)
    print(f"\nArquivo '{csv_file}' gerado e pronto para download.")

except NameError:
    print("Erro: O widget 'widget' no foi encontrado. Certifique-se de que a clula com o widget foi executada.")
except Exception as e:
    print(f"Erro ao processar ou baixar os dados do widget: {e}")
```

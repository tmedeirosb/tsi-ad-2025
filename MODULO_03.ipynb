## Tipos de Variáveis

Em estatística, as variáveis podem ser classificadas em diferentes tipos, o que influencia a forma como as analisamos e visualizamos. Os principais tipos são:

- **Variáveis Quantitativas:** Representam quantidades numéricas.
    - **Contínuas:** Podem assumir qualquer valor dentro de um intervalo (ex: altura, peso, temperatura).
    - **Discretas:** Podem assumir apenas valores inteiros (ex: número de filhos, número de carros).

- **Variáveis Qualitativas:** Representam categorias ou atributos.
    - **Nominais:** Não possuem uma ordem natural (ex: cor, gênero, estado civil).
    - **Ordinais:** Possuem uma ordem natural (ex: escolaridade - fundamental, médio, superior; classe social - baixa, média, alta).

Vamos identificar os tipos de variáveis no nosso conjunto de dados. Podemos usar o método `.info()` para ter uma ideia dos tipos de dados que o pandas inferiu.

```python
import pandas as pd

# exibir todas as colunas
pd.set_option('display.max_columns', None)

url = "https://raw.githubusercontent.com/tmedeirosb/tsi-ad-2025/refs/heads/main/dados/dados_workflow_ivan.csv"
df = pd.read_csv(url)

display(df.head())

display(df.info())
```

### Tabelas de Frequência e Estatísticas Descritivas

Agora, vamos gerar tabelas de frequência e estatísticas descritivas para algumas variáveis, demonstrando como analisar diferentes tipos.

**Variáveis Qualitativas: Tabela de Frequência**

Para variáveis qualitativas, a tabela de frequência nos mostra a contagem e a proporção de cada categoria. Vamos usar a coluna `descricao` como exemplo, que indica o status do aluno (Matriculado, Cancelado, etc.).

```python
# Tabela de frequências absolutas (contagem) da variável 'descricao'
contagem_descricao = df['descricao'].value_counts()

# Tabela de frequências relativas (%) da variável 'descricao'
percent_descricao = df['descricao'].value_counts(normalize=True) * 100

# Combinar as duas tabelas
tabela_frequencia_descricao = pd.concat([contagem_descricao, percent_descricao], axis=1, keys=['Contagem', '%'])

display(tabela_frequencia_descricao)
```

**Variáveis Quantitativas: Estatísticas Descritivas**

Para variáveis quantitativas, podemos usar o método `.describe()` para obter um resumo das principais estatísticas, como média, mediana, desvio padrão, mínimo, máximo e quartis. Vamos usar a coluna `idade` como exemplo.

```python
# Estatísticas descritivas da variável 'idade'
estatisticas_idade = df['idade'].describe()

display(estatisticas_idade)
```

### Visualização de Variáveis

Visualizar os dados é uma etapa crucial na análise exploratória. O tipo de gráfico que usamos depende do tipo de variável que estamos analisando.

**Variáveis Qualitativas: Gráficos de Barras**

Para variáveis qualitativas, gráficos de barras são ideais para mostrar a frequência de cada categoria. Vamos criar um gráfico de barras para a variável `descricao`.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='descricao', order=df['descricao'].value_counts().index)
plt.title('Distribuição da Variável Descrição')
plt.xlabel('Contagem')
plt.ylabel('Descrição')
plt.show()
```

**Variáveis Quantitativas: Histogramas**

Para variáveis quantitativas, histogramas nos ajudam a visualizar a distribuição dos valores. Vamos criar um histograma para a variável `idade`.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='idade', kde=True)
plt.title('Distribuição da Variável Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='idade', bins=40, stat='density') # Adicionado stat='density'
plt.title('Distribuição da Variável Idade (Densidade Normalizada)')
plt.xlabel('Idade')
plt.ylabel('Densidade') # Rótulo alterado para Densidade
plt.show()
```

### Medidas de Tendência Central

As medidas de tendência central nos dão uma ideia do valor típico ou central de um conjunto de dados. As mais comuns são:

- **Média:** A soma de todos os valores dividida pelo número de valores. É sensível a valores extremos (outliers).
- **Mediana:** O valor do meio em um conjunto de dados ordenado. É menos sensível a outliers do que a média.
- **Moda:** O valor que aparece com mais frequência no conjunto de dados. É a única medida de tendência central que pode ser usada com dados nominais.

No resultado do método `.describe()` que vimos para a variável `idade`, podemos identificar a média (`mean`) e a mediana (que é o 50º percentil, `50%`). A moda não é diretamente fornecida pelo `.describe()`, mas podemos obtê-la usando o método `.mode()`.

```python
# Calcular a moda da variável 'idade'
moda_idade = df['idade'].mode()

print(f"Média da idade: {estatisticas_idade['mean']:.2f}")
print(f"Mediana da idade: {estatisticas_idade['50%']:.2f}")
print(f"Moda da idade: {moda_idade.to_list()}")
```

### Assimetria

A assimetria nos diz sobre a simetria de uma distribuição de dados.

- **Distribuição Simétrica:** Os dados são distribuídos uniformemente em ambos os lados do centro. A média, mediana e moda são aproximadamente iguais. Um exemplo clássico é a distribuição normal.
- **Distribuição Assimétrica Positiva (à direita):** A cauda da distribuição é mais longa no lado direito. A média geralmente é maior que a mediana, que por sua vez é maior que a moda.
- **Distribuição Assimétrica Negativa (à esquerda):** A cauda da distribuição é mais longa no lado esquerdo. A média geralmente é menor que a mediana, que por sua vez é menor que a moda.

Vamos visualizar um exemplo de distribuição simétrica e um exemplo de distribuição assimétrica.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Gerar dados para uma distribuição simétrica (Normal)
dados_simetricos = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(10, 6))
sns.histplot(dados_simetricos, kde=True, bins=30)
plt.title('Exemplo de Distribuição Simétrica (Normal)')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()

# Calcular e imprimir média, mediana e moda para comparação
media_simetrica = np.mean(dados_simetricos)
mediana_simetrica = np.median(dados_simetricos)
# A moda para dados contínuos não é bem definida, mas podemos aproximar usando bins ou densidade
# Para simplificar, vamos apenas imprimir a média e a mediana aqui.
print(f"Média (Simétrica): {media_simetrica:.2f}")
print(f"Mediana (Simétrica): {mediana_simetrica:.2f}")
```

```python
# Gerar dados para uma distribuição assimétrica (Log-Normal - assimétrica positiva)
dados_assimetricos = np.random.lognormal(mean=0, sigma=1, size=1000)

plt.figure(figsize=(10, 6))
sns.histplot(dados_assimetricos, kde=True, bins=30)
plt.title('Exemplo de Distribuição Assimétrica Positiva (Log-Normal)')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()

# Calcular e imprimir média, mediana e moda para comparação
media_assimetrica = np.mean(dados_assimetricos)
mediana_assimetrica = np.median(dados_assimetricos)
# Moda para dados contínuos - aproximada
from scipy.stats import gaussian_kde
kde = gaussian_kde(dados_assimetricos)
x_grid = np.linspace(dados_assimetricos.min(), dados_assimetricos.max(), 1000)
moda_assimetrica = x_grid[np.argmax(kde.evaluate(x_grid))]

print(f"Média (Assimétrica): {media_assimetrica:.2f}")
print(f"Mediana (Assimétrica): {mediana_assimetrica:.2f}")
print(f"Moda (Assimétrica - Aproximada): {moda_assimetrica:.2f}")
```

### Medidas Separatrizes

As medidas separatrizes dividem um conjunto de dados ordenado em partes iguais. As mais comuns são:

- **Quartis:** Dividem os dados em quatro partes iguais.
    - Primeiro Quartil (Q1): 25% dos dados são menores ou iguais a este valor.
    - Segundo Quartil (Q2): 50% dos dados são menores ou iguais a este valor (equivalente à Mediana).
    - Terceiro Quartil (Q3): 75% dos dados são menores ou iguais a este valor.

- **Percentis:** Dividem os dados em cem partes iguais. O p-ésimo percentil é o valor abaixo do qual p% dos dados caem.

Podemos obter os quartis diretamente do resultado do método `.describe()`, que vimos anteriormente para a variável `idade` (25%, 50%, 75%). Para outros percentis, podemos usar o método `.quantile()`.

```python
# Exemplo: Calcular o 90º percentil da variável 'idade'
percentil_90_idade = df['idade'].quantile(0.90)

print(f"Primeiro Quartil (Q1) da idade: {estatisticas_idade['25%']:.2f}")
print(f"Terceiro Quartil (Q3) da idade: {estatisticas_idade['75%']:.2f}")
print(f"90º Percentil da idade: {percentil_90_idade:.2f}")
```

### Medidas de Dispersão

As medidas de dispersão indicam o quão espalhados ou variados os dados estão em torno de uma medida de tendência central. As mais comuns incluem:

- **Amplitude Total:** A diferença entre o valor máximo e o valor mínimo. É simples, mas muito sensível a outliers.
- **Variância:** A média dos quadrados das diferenças de cada valor em relação à média.
- **Desvio Padrão:** A raiz quadrada da variância. É a medida de dispersão mais comum e está na mesma unidade dos dados originais.
- **Intervalo Interquartil (IIQ):** A diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1). Menos sensível a outliers do que a amplitude total.

Podemos obter o desvio padrão (`std`), o valor mínimo (`min`) e o valor máximo (`max`) a partir do método `.describe()`. O Intervalo Interquartil pode ser calculado a partir dos valores de Q1 e Q3 obtidos no `.describe()`.

Além dessas medidas numéricas, visualizações como Box Plots e Violin Plots são excelentes para entender a dispersão e a distribuição dos dados graficamente.

**Box Plot (Diagrama de Caixa)**

O Box Plot mostra a distribuição dos dados através de seus quartis. A "caixa" representa o Intervalo Interquartil (IIQ), com a linha no meio indicando a mediana (Q2). Os "bigodes" (whiskers) geralmente se estendem até 1.5 vezes o IIQ a partir dos quartis, e pontos fora dessa faixa são considerados potenciais outliers.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='idade')
plt.axvline(estatisticas_idade['50%'], color='red', linestyle='--', label=f'Mediana: {estatisticas_idade["50%"]:.2f}')
plt.title('Box Plot da Variável Idade com Mediana Destacada')
plt.xlabel('Idade')
plt.legend()
plt.show()
```

**Violin Plot**

O Violin Plot combina o Box Plot com um gráfico de densidade de probabilidade (Kernel Density Estimate - KDE) rotacionado. Ele mostra a distribuição dos dados de forma mais completa do que o Box Plot, exibindo a densidade dos dados em diferentes valores. A parte mais larga do "violino" indica onde a densidade de dados é maior.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='idade')
plt.title('Violin Plot da Variável Idade')
plt.xlabel('Idade')
plt.show()
```

### Outliers

Outliers são observações que se afastam significativamente dos demais dados em um conjunto. Eles podem ocorrer devido a erros de medição, erros de entrada de dados ou podem representar variações genuínas no processo que está sendo estudado. A identificação e o tratamento de outliers são importantes porque eles podem distorcer medidas estatísticas (como a média) e impactar a performance de modelos de machine learning.

Como vimos no Box Plot da variável `idade`, os pontos individuais que aparecem fora dos "bigodes" são considerados potenciais outliers de acordo com o critério de 1.5 vezes o Intervalo Interquartil (IIQ).

Existem métodos mais formais para identificar outliers, mas visualizações como Box Plots e Violin Plots são um excelente primeiro passo para detectá-los.
